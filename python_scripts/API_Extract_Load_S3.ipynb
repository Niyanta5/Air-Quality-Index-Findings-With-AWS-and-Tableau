```python
def format_state_county(state_county_list):
  """Formats a list of state_county tuples (integers) with leading zeros.

  Args:
      state_county_list: A list of tuples containing state code (integer) and county code (integer).

  Returns:
      A list of tuples containing the formatted state code (string) and county code (string).
  """

  return [(format_state_code(state_county), format_county_code(state_county)) for state_county in state_county_list]

def format_state_code(state_county):
    """Formats state code (integer) with a leading zero.

    Args:
        state_code: An integer representing the state code.

    Returns:
        A string containing the formatted state code with a leading zero.
    """
    state_code, county_code = state_county
    if state_code < 10:
        return f"0{state_code}"
    return str(state_code)

def format_county_code(state_county):
    """Formats county code (integer) with leading zeros.

    Args:
        county_code: An integer representing the county code.

    Returns:
        A string containing the formatted county code with leading zeros.
    """
    state_code, county_code = state_county
    if county_code < 10:
        return f"00{county_code}"
    elif county_code < 100:
        return f"0{county_code}"
    else:
        return str(county_code)


formatted_list = format_state_county(state_county)

print(formatted_list)
```
